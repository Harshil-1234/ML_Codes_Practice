# 4. Sep Parameter

```python
#by default separator in read_csv is ',' so while reading tsv file, it will no read any comma and give wrong output
pd.read_csv('movie_titles_metadata.tsv',sep = '\t',names=['Sno','name','release_year','rating','votes','genre'])
```

# 5. Index_col parameter

```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

```python
#header tells from which row data will start, header = 1 means make 1st row column name and start from 2nd row
pd.read_csv('aug_train.csv',header=3)
```

# 6. usecols param (allows to display specific columns only

```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

# 7. Squeeze Parameter
If we want only one column, we can set squeeze = true, then it will return a pandas SERIES instead of dataframes
<br>pd.read_csv('aug_train.csv',usecols=['enrollee_id'],squeeze = TRUE)

# 8. Skiprows
allows to skip rows and display remaining data <br>
pd.read_csv('aug_train.csv',usecols=['enrollee_id'],skiprows = [0,1]) will skip row 0 and 1.

# 9. nrows Parameter
pd.read_csv('aug_train.csv',nrows=100) print first 100 rows only
